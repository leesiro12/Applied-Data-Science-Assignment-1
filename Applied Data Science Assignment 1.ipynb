```python
#plot the employment of the whole df on a line plot
def plot(df):
    
    plt.figure(dpi=144)
    
    df.reset_index().plot(x="year",y="employment")

    plt.xlabel('Year')
    plt.ylabel('Employment')
    plt.legend()
    plt.show()
    return
```

```python
def plot_correlation(df_noNaN, method):
    """
    Plots correlation of employment and companies with different methods
    """
    fig, ax = plt.subplots(dpi=144)

    mask = np.triu(np.ones_like(df_noNaN.corr()))
    sns.heatmap(df_noNaN.corr(method=method), ax=ax, vmin=-1, vmax=1, cmap='RdBu', annot=True, mask=mask)
    plt.title(method.capitalize())
    plt.show()
    return
```

```python
plot_correlation(df_noNaN, 'pearson')
plot_correlation(df_noNaN, 'kendall')
```
